# ***Ordinal Encoding***

```python
data_dict={'Fridge_Company':['LG','Samsung','Voltas','Whirlpool','Godrej','Haier'],'Review':['Poor','Average','Good','Very Good','Excellent','Good']}
```

```python
df=pd.DataFrame(data_dict)
df
```

```python
new_df=pd.DataFrame(0,index=df.index,columns=['Review_Ordinal'])
```

```python
new_df
```

```python
ordinal_encode_df=pd.concat([df,new_df],axis=1)
```

```python
ordinal_encode_df
```

```python
j=1
map_dict={}
for i in df['Review'].unique():
  map_dict[i]=j
  j+=1
```

```python
map_dict
```

```python
i=0
for r in df['Review']:
  ordinal_encode_df.loc[i,'Review_Ordinal']=map_dict[r]
  i+=1
```

```python
ordinal_encode_df
```

# ***Imputaion***

##Mean

```python
import numpy as np
```

```python
df=pd.DataFrame({
    'Name':['Ram','Raju','John','Shyam','Yuvi'],
    'Age':[25,40,np.nan,35,np.nan]
})
```

```python
df
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df
```

##Median

```python
import numpy as np
```

```python
df=pd.DataFrame({
    'Name':['Ram','Raju','John','Shyam','Yuvi'],
    'Age':[25,40,np.nan,35,np.nan]
})
```

```python
df
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df
```

##Mode

```python
import numpy as np
```

```python
df=pd.DataFrame({
    'Name':['Ram','Raju','John','Shyam','Yuvi','Abhinav'],
    'Age':[25,40,np.nan,35,np.nan,25]
})
```

```python
df
```

```python
df['Age']=df['Age'].fillna(df['Age'].mode()[0])
```
